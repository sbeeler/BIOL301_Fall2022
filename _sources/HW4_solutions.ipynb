## Problem 1

### (a)

Fixed points occur when:

\begin{align}
0 &= rx + x^2 \\[1em]
0 &= x(r+x).
\end{align}

So the fixed points are $x=0$ and $x=-r$.

### (b)

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=[20,5], sharey=True)

x = np.linspace(-1.1,1.1,100)

r = -1
y = r*x + x**2
ax1.plot(x, y)
ax1.axhline(y=0, color="gray")
ax1.set(xlabel='x', ylabel='$\dot x$', title="r < 0")

r = 0
y = r*x + x**2
ax2.plot(x, y)
ax2.axhline(y=0, color="gray")
ax2.set(xlabel='x',  title="r = 0")

r = 1
y = r*x + x**2
ax3.plot(x, y)
ax3.axhline(y=0, color="gray")
ax3.set(xlabel='x',  title="r > 0");
```

## (c)

When $r < 0$:

- The fixed point at $x=0$ is stable
- The other fixed point at $x=r$ unstable

When $r = 0$:

- There is only a fixed point at $x=0$, which is semi-stable

When $r > 0$:

- The fixed point at $x=0$ is now unstable
- The other fixed point at $x=r$ is now stable

## Problem 2

### (a)

A fixed point will occur when

$$ 0 = x^2 + rx + 1.$$

By using the quadratic formula (with $a=1$, $b=r$, and $c=1$), we find that this equation will be satisfied when 

$$ x^{\ast} = \frac{-r \pm \sqrt{r^2 -4}}{2}.$$

We see that there are generally two fixed points given that this is a quadratic equation. However, when $r=2$ or $r=-2$, $ \sqrt{r^2 -4}$ evaluates to zero, resulting in only a single fixed point. 

Additionally, when $-2 < r < 2$, $ \sqrt{r^2 -4}$ will evaluates to a an imaginary number, meaning there are no fixed points in this range.

### (b)

```python
# set up 2 by 3 grid of plots
fig, axs = plt.subplots(2, 3, figsize=(20,10), sharey=True, sharex=True)

x = np.linspace(-3,3,100)
rs = [-3, -2, 0, 2, 3]

# loop through plots and values of r
for ax, r in zip(axs.flat, rs):
    growth = 1 + r*x
    decay = -x**2
    ax.plot(x, growth)
    ax.plot(x, decay)
    ax.axvline(x=0, color="gray")
    ax.axhline(y=0, color="gray")
    ax.set(xlabel='x', ylabel='compoent of $\dot x$', title= "r = "+str(r))

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

### (d)

The bifurcation will occur when

\begin{align}
\frac{d}{dx} 1 + rx &= \frac{d}{dx} - x^2 \\[1em]
r &= -2x
\end{align}

Let's plug this value of $r$ into our condition for a fixed point,

\begin{align}
1 + rx &= -x^2 \\[1em]
1 + (-2x)x &= -x^2 \\[1em]
1 - 2x^2 &= -x^2 \\[1em]
1 &= x^2 \\[1em]
x &= \pm 1 \\[1em]
\end{align}

Lastly, plugging in $x= \pm 1$ into $r=-2x$, we find that the bifurcations should occur at $r = \pm 2$. This matches exactly what we saw in the previous parts, where $r<|2|$ had no fixed points, but $r>|2|$ had two fixed points. Thus, $r=|2|$ is where this bifurcation, or change in the number of fixed points occurs.

## Problem 3

### (a)

In the first reaction, the chemical species A and X combine in some way to make another unit of X. In this reaction, A is consumed and a unit of X is produced. 

In the second reaction, the chemical species X and B combine to make a third chemical species, C. In this reaction, X and B are consumed and C is produced.

### (b)

- $k_1ax$ corresponds to the forward reaction of the first chemical reaction, in which X is produced by the interaction of A and X with rate $k_1$. This term is positive because it relates to the production of X. 

- $k_{-1}x^2$ corresponds to the reverse reaction of the first chemical reaction, in which 2 units of X revert back to a unit of A and a unit of X with rate $k_{-1}$. This term is negative as it relates to an overall loss of X.

- Lastly, $k_2bx$ corresponds to the second chemical reaction, in which B and X interact to form C with rate $k_2$. This term is negative because it also relates to a loss of X.

### (c)

$$\dot x = (k_1a - k_2b)x - k_{-1}x^2$$

### (d)

When can use linear stability analysis to determine the stability of the fixed point at $x=0$. Let's first take the derivative of $\dot x$:

$$\ddot x = (k_1a - k_2b) - 2k_{-1}x.$$

When we plug in the fixed point $x=0$, we get:

$$\ddot x = k_1a - k_2b. $$

We see that $\ddot x$ will be negative, corresponding to a stable fixed point when $k_2b > k_1a$. This means that the second reaction will out pace the first forward reaction, causing the amount of X to go to zero. This is because the second reaction depletes X while the first reaction creates X. If depletion is occurring more quickly than production, than X will never have a chance to accumulate, and the amount will be stuck a 0.

### (b)

The bifurcation will occur when $x(x-1) = h$ and 

\begin{align}
\frac{d}{dx} x(1-x) &= \frac{d}{dx} h \\[1em]
\frac{d}{dx} x-x^2 &= \frac{d}{dx} h \\[1em]
1 - 2x &= 0 \\[1em]
1 &= 2x \\[1em]
\frac{1}{2} &= x \\[1em]
\end{align}

Plugging $x=1/2$ into $x(1-x)=h$, we get

\begin{align}
x(1-x) &= h \\[1em]
\frac{1}{2}\left( 1 - \frac{1}{2} \right) &= h \\[1em]
\frac{1}{4} &=  h
\end{align}

We see that the bifurcation occurs when $h_c=1/4$, which matches what we see in the plots above.

### (c)

When $h > h_c$, the population will be harvested to extinction because the amount harvested always exceeds the amount of population growth.

When $h < h_c$, the population will settle on the stable fixed point found on the right of the two fixed points.
